```python
import numpy as np

ims = np.load('sprites_1788_16x16.npy')
```

```python
from PIL import Image
from einops import rearrange

rng = np.random.default_rng()

Image.fromarray(
    rearrange(
        rng.choice(ims, size=(25,), replace=False),
        "(b1 b2) h w c -> (b1 h) (b2 w) c",
        b1=5,
    )
).resize((128 * 5, 128 * 5), Image.NEAREST)
```

```python
from torch.utils.data import Dataset, DataLoader

class SpritesDataset(Dataset):
    def __init__(self):
        self.ims = np.load('sprites_1788_16x16.npy')

    def __len__(self):
        return len(self.ims)

    def __getitem__(self, idx):
        return self.ims[idx]
        
```
